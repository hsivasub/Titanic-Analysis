## Travelling with Family or Travelling Alone 

```python
titanic_df.head()
```

## Creating a Travel column which reads either 'Alone' or 'Family'

```python
def alone_family(total):
    sib, parent= total
    if (sib+parent>0):
        return 'Family'
    else:
        return 'Alone'
```

```python
titanic_df['Travel'] = titanic_df[['SibSp','Parch']].apply(alone_family,axis=1)
titanic_df.head()
```

```python
survive1= titanic_df[titanic_df.Survived==1]
survive1.head()
```

```python
sns.countplot('Travel', data=titanic_df, hue= 'Pclass',order=['Family','Alone'])
sns.countplot('Travel', data=survive1, hue= 'Pclass', saturation=0.1, palette='hot',order=['Family','Alone'])
```

### Here we can see the people who came with Family had more chance of Survival.
